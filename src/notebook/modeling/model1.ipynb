```python
import re

def extract_ram_size(ram_str):
    match = re.search(r'\b(\d+)\s*GB\b', str(ram_str))
    if match:
        return int(match.group(1))
    else:
        return None

df_filtered["RAM"] = df_filtered["RAM"].apply(extract_ram_size)
df_filtered["RAM"].value_counts()
```

```python
import re

def extract_monitor_size(monitor_str):
    match = re.search(r'\b(\d+(\.\d+)?)\s*Inches?\b', str(monitor_str))
    if match:
        return float(match.group(1))
    else:
        return None

df_filtered["Monitor"] = df_filtered["Monitor"].apply(extract_monitor_size)

df_filtered["Monitor"].value_counts()
```

```python
df_filtered["Resolution"].value_counts().index
```

```python
def extract_resolution(resolution_str, dimension='width'):
    match = re.search(r'(\d+)\s*[xX]\s*(\d+)', str(resolution_str))
    if match:
        if dimension.lower() == 'width':
            return int(match.group(1))
        elif dimension.lower() == 'height':
            return int(match.group(2))
    return None

df_filtered["Width"] = df_filtered["Resolution"].apply(extract_resolution, dimension='width')
df_filtered["Height"] = df_filtered["Resolution"].apply(extract_resolution, dimension='height')
```

```python
def extract_storage_capacity(storage_str):
    match = re.search(r'(\d+)\s*([TtGgMmBb]?)B?\s*(?:\+\s*(\d+)\s*([TtGgMmBb]?)B?)?', str(storage_str))
    
    if match:
        size1, unit1, size2, unit2 = match.groups()

        size1 = convert_to_gb(int(size1), unit1)
        size2 = convert_to_gb(int(size2), unit2) if size2 else None

        return size1, size2

    return None, None

def convert_to_gb(size, unit):
    conversion_factors = {'T': 1024**3, 'G': 1024**2, 'M': 1024, 'B': 1}
    return size * conversion_factors.get(unit.upper(), 1)

df_filtered["Storage_Capacity1"], _ = zip(*df_filtered["Storage"].apply(extract_storage_capacity))
df_filtered["Storage_Capacity1"] = df_filtered["Storage_Capacity1"].apply(lambda x: int(x / 1024**2))
df_filtered = df_filtered[df_filtered["Storage_Capacity1"] > 0]
df_filtered["Storage_Capacity1"].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVR
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
import joblib
import warnings
warnings.filterwarnings("ignore")
```

```python
X, y = limited_df.drop("Price", axis=1), limited_df["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=28)

scaler = StandardScaler()
X_train_scalered = scaler.fit_transform(X_train)
X_test_scalered = scaler.fit_transform(X_test)
```

```python
def train_model(X_train, y_train, model_name: str, grid_search=False):
    if model_name=='LinearSVR':
        if grid_search==True:
            param_grid = {
                'C': [0.01, 0.1, 1, 10, 100, 200],
                'epsilon': [0.01, 0.1, 1, 10],
                'max_iter': [100, 500, 800, 1000],
                'tol': [1e-4, 1e-3, 1e-2]
            }

            svr = LinearSVR()
            grid_search = GridSearchCV(svr, param_grid, scoring='neg_mean_squared_error')
            grid_search.fit(X_train, y_train)

            print("Best Params:", grid_search.best_params_)
            print("Best Score:", -grid_search.best_score_)

            return grid_search.best_estimator_
        
        else:
            svr = LinearSVR()
            svr.fit(X_train_scalered, y_train)

            return svr
    
    if model_name=='SVR':
        if grid_search==True:
            param_grid = {
                'C': [0.01, 0.1, 1, 10, 100],
                'epsilon': [0.01, 0.1, 1, 10],
                'gamma': ['scale', 'auto', 0.1, 0.01, 0.001],
                'kernel': ['rbf']
            }

            svr = SVR()
            grid_search = GridSearchCV(svr, param_grid, scoring='neg_mean_squared_error')
            grid_search.fit(X_train, y_train)

            print("Best Params:", grid_search.best_params_)
            print("Best Score:", -grid_search.best_score_)

            return grid_search.best_estimator_
        
        else:
            svr = SVR(kernel='linear')
            svr.fit(X_train, y_train)

            return svr
    
    if model_name=='KNN':
        if grid_search==True:
            param_grid = {
                'n_neighbors': [3, 5, 7, 9, 11],
                'weights': ['uniform', 'distance'],
                'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
            }

            knn = KNeighborsRegressor()
            grid_search = GridSearchCV(knn, param_grid, scoring='neg_mean_squared_error')
            grid_search.fit(X_train, y_train)

            print("Best Params:", grid_search.best_params_)
            print("Best Score:", -grid_search.best_score_)

            return grid_search.best_estimator_
        
        else:
            knn = KNeighborsRegressor()
            knn.fit(X_train, y_train)

            return knn
```

```python
def predict(model: LinearSVR, X_test, y_test):
    y_pred = model.predict(X_test_scalered)
    print(mean_squared_error(y_test, y_pred))

    plt.figure(figsize=(16,16))
    plt.scatter(y_pred, y_test)
    plt.plot(range(0,5000), range(0,5000), c='red')
    plt.show()
```

```python
import numpy as np
```

```python
def evaluation(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    
    mape = mean_absolute_percentage_error(y_test, y_pred)
    print(f'Mean Squared Error: {mse:.2f}')
    print(f'Mean Absolute Percentage Error: {mape:.2f}%')

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
def save_model(model, model_name):
    filename = f"{model_name}_model.pkl"
    joblib.dump(model, filename)
    print(f"Model saved as {filename}")
```

```python
MODEL_NAME = 'KNN'
GRID_SEARCH = False

trained_model = train_model(X_train=X_train_scalered, y_train=y_train, model_name=MODEL_NAME, grid_search=GRID_SEARCH)

predict(model=trained_model, X_test=X_test_scalered, y_test=y_test)
evaluation(model=trained_model, X_test=X_test_scalered, y_test=y_test)
    
```
