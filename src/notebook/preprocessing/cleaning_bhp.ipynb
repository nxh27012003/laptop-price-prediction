```python
df_copy = df
df_copy['Operating-System'] = df_copy['Operating-System'].replace({
    'Windows 11 Pro (64-Bit)': 'Windows 11 Pro',
    'Windows 11 Pro for Workstations (64-Bit)': 'Windows 11 Pro',

    'Windows 11 Home (64-Bit)': 'Windows 11 Home',
    'Windows 11 Home in S Mode': 'Windows 11 Home',

    'Windows 10 Pro (64-Bit)': 'Windows 10 Pro',
    'Windows 10 Pro Education': 'Windows 10 Pro',

    'Windows 10 Home in S Mode': 'Windows 10 Home',
    'Windows 11 SE (64-Bit)': 'Windows 10 Home',
    'Windows 10 IoT Enterprise (64-Bit)': 'Windows 10 Home',
    'Dell ThinOS (64-Bit)': 'Windows 10 Home',

    'Chrome OS Enterprise': 'Chrome OS',
    'Chrome OS (64-Bit)': 'Chrome OS',
})

df_copy.loc[df_copy['Operating-System'].isna() & df_copy['Name'].str.startswith('Apple'), 'Operating-System'] = 'macOS'
df_copy.loc[df_copy['Operating-System'].isna() & ~df_copy['Name'].str.startswith('Apple'), 'Operating-System'] = 'Windows 11 Home'

df_copy['Operating-System'].value_counts()
```

```python
df_copy = df_copy.drop(labels=["Box-Dimension (LxWxH)", "Dimensions", "Power-Supply", "Keyboard", "GPS",
                                "Celluar-Support", "Media/Memory Card Slot", "Built-In-Microphones", 
                                "Built-In-Speakers", "Audio-I/O", "Variable-Fresh-Technology", "Contrast-Ratio",
                                "Finish", "Aspect-Ratio", "Column-Gamut", "vPro-Support", "ECC-Memory", "Model-Year", 
                                "WiFi", "Display-I/O", "Inputs/Outputs"], axis=1)
df_copy["Brand"] = df_copy["Name"].str.split().apply(lambda x: x[0])
```

```python
df_copy = df_copy.drop("Name", axis=1)
```

```python
df_copy["Total-Installed-Memory"].fillna(value="8 GB", inplace=True)
df_copy["Total-Installed-Memory"] = df_copy["Total-Installed-Memory"].str.split().apply(lambda x: x[0]).astype(int)
```

```python

#RAM
df_copy["Total-Installed-Memory"] = df_copy["Total-Installed-Memory"].str.split().apply(lambda x: x[0]).astype(int)
#Storage
def convert_memory(amount: str) -> int:
    value_, type_ = amount.split()
    if type_ == "TB":
        return int(value_) * 1024
    
    if type_ == "GB":
        return int(value_)
df_copy["Total-Installed-Capacity"] = df_copy["Total-Installed-Capacity"].apply(convert_memory)
```

```python
df_copy["Display-Size"].fillna("14\"", inplace=True)
df_copy["Display-Size"] = df_copy["Display-Size"].apply(lambda x: x[:-1]).astype(float)
```

```python
df_copy["Resolution"].fillna("1920 x 1080", inplace=True)
df_copy["Width"] = df_copy["Resolution"].str.split().apply(lambda x: x[0]).astype(int)
df_copy["Height"] = df_copy["Resolution"].str.split().apply(lambda x: x[-1]).astype(int)
df_copy = df_copy.drop("Resolution", axis=1)
```

```python
df_copy["Touch-Screen"].fillna("No", inplace=True)
df_copy["Touch-Screen"] = df_copy["Touch-Screen"].replace({
    "No": 0,
    "Yes": 1
})
```

```python
def convert_memory(amount: str) -> int:
    value_, type_ = amount.split()
    if type_ == "TB":
        return int(value_) * 1024
    
    if type_ == "GB":
        return int(value_)
df_copy["Total-Installed-Capacity"] = df_copy["Total-Installed-Capacity"].apply(convert_memory)
```

```python
df_copy["Total-Installed-Capacity"].fillna("512 GB", inplace=True)
df_copy["Total-Installed-Capacity"] = df_copy["Total-Installed-Capacity"].apply(convert_memory)
```

```python
df_copy["Bluetooth"].fillna("5.0", inplace=True)
df_copy["Bluetooth"] = df_copy["Bluetooth"].replace({
    "Yes": "5.1",
    "No": "4.2"
})
df_copy["Bluetooth"] = df_copy["Bluetooth"].astype(float)
```

```python
df_copy['Weight'] = df_copy['Weight'].str.split('/').str[1].str.strip().str.rstrip('kg').str.strip().astype(float)
mean_weight = df_copy['Weight'].mean()
df_copy['Weight'] = df_copy['Weight'].fillna(mean_weight)
```

```python
df_copy["L3-Cache"].fillna("12 MB", inplace=True)
df_copy['L3-Cache'] = df_copy['L3-Cache'].replace('Integrated', 0)
df_copy['L3-Cache'] = df_copy['L3-Cache'].replace({' MB': ''}, regex=True).astype(float)
```

```python
df_copy["Graphics-Type"].fillna("Integrated", inplace=True)
df_copy["Graphics-Type"].value_counts()
df_copy = pd.get_dummies(df_copy, columns=["Graphics-Type"])
```

```python
df_copy.drop("Memory-Type", axis=1, inplace=True)
```

```python
df_copy.drop("Memory-Configuration", axis=1, inplace=True)
```

```python
df_copy.drop("Maximum-Memory-Capacity", axis=1, inplace=True)
```

```python
df_copy["Memory-Slots"].value_counts()
```

```python
df_copy["Memory-Slots"].fillna("1", inplace=True)
df_copy["Memory-Slots"] = df_copy["Memory-Slots"].replace({
    "2 (1 Available)": "2",
    "2 (0 Available)": "2",

    "4 (2 Available)": "4",
    "4 (0 Available)": "4",

    "1 (0 Available)": "1",
    "1 (1 Available)": "1"
})
```

```python
df_copy["Memory-Slots"] = df_copy["Memory-Slots"].astype(int)
df_copy["Memory-Slots"].value_counts()
```

```python
df_copy["Panel-Type"].fillna("IPS", inplace=True)
df_copy["Panel-Type"] = df_copy["Panel-Type"].replace({
    "IPS-Type LCD": "IPS",
    "Mini-LED LCD": "IPS",

    "TN LCD": "TN",
    "VA LCD": "VA",
    "TFT LCD": "VA"
})
df_copy = pd.get_dummies(df_copy, columns=["Panel-Type"])
```

```python
df_copy["Refresh-Rate"].fillna("120 Hz", inplace=True)
df_copy["Refresh-Rate"] = df_copy["Refresh-Rate"].str.split().apply(lambda x: x[0])
df_copy["Refresh-Rate"] = df_copy["Refresh-Rate"].astype(int)
```

```python
df_copy.drop("Response-Time", axis=1, inplace=True)
```

```python
df_copy.drop("Storage-Drive-Type", axis=1, inplace=True)
```

```python
df_copy["Maximum-Brightness"].fillna("250 nits / cd/m2", inplace=True)
df_copy["Maximum-Brightness"] = df_copy["Maximum-Brightness"].str.split().apply(lambda x: x[0]).astype(int)
```

```python
df_copy["Storage-Expansion"].fillna("No", inplace=True)
df_copy["Storage-Expansion"] = df_copy["Storage-Expansion"].replace({"No": 0, ".*": 1}, regex=True)

df_copy["Storage-Expansion"].value_counts()
```

```python
df_copy.drop("Network-I/O", axis=1, inplace=True)
```

```python
df_copy.drop("Webcam", axis=1, inplace=True)
```

```python
df_copy["Batery-Chemistry"].fillna("Lithium-Ion Polymer (LiPo)", inplace=True)
df_copy["Batery-Chemistry"].replace("Lithium-Ion Phosphate (LiFe)", "Lithium-Ion", inplace=True)
df_copy["Batery-Chemistry"].value_counts()
```

```python
df_copy = pd.get_dummies(df_copy, columns=["Batery-Chemistry"])
```

```python
df_copy["Capacity"] = df_copy["Capacity"].str.extract(r'(\d+\.\d+)').astype(float)
df_copy["Capacity"].fillna(df_copy["Capacity"].mean(), inplace=True)
df_copy["Capacity"].value_counts()
```

```python
df_copy.drop("Operating-Tempurature", axis=1, inplace=True)
```

```python
df_copy.drop("Operating-Humidity", axis=1, inplace=True)
```

```python
df_copy.drop("Package-Weight", axis=1, inplace=True)
```

```python
df_copy["Price"] = df_copy["Price"].replace('[\$,]', '', regex=True).astype(float)
df_copy["Price"].fillna(df_copy["Price"].mean(), inplace=True)

df_copy["Price"].value_counts()
```

```python
df_copy["Brand"] = df_copy["Brand"].replace({
    "Samsung": "Other",
    "Razer": "Other",
    "Gigabyte": "Other",
    "VAIO": "Other",
    "Core": "Other",
    "Dynabook": "Other",
    "XPG": "Other"
})
df_copy["Brand"].value_counts()
```

```python
def fill_processor(row):
    if pd.isna(row["Processor"]):
        if row["Brand_Apple"] == 1:
            return "Apple M2"
        else:
            return "Intel Core i5-1335U (13th Gen)"
    else:
        return row["Processor"]

df_copy["Processor_"] = df_copy.apply(fill_processor, axis=1)
```
